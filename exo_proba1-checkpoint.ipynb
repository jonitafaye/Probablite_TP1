```python
# SIMULATION DE LA LOI BINOMIALE
from random import random
from scipy.special import binom
#N=input("entrer le nombre N")
#n=input("entrer le nombre n")
#p=input("entrer le nombre p")
def binomial(N,n,p):
 s=(n+1)*[0];e=(n+1)*[0]
 for i in range(N):
     c=0
 for j in range(n):
     x=random()
 if x<p:
     c=c+1
     s[c]=s[c]+1
 for i in range(n+1):
     s[i]=s[i]/N
 for i in range(n+1):
     e[i]=binom(n,i)*p**i*(1-p)**(n-i)
 return(s,e) 
binomial(10,1,2)
```

```python
#REPRESENTATION GRAPHIQUE DE LA LOI BINOMIAL 
from random import random
from scipy.special import binom
import matplotlib.pyplot as plt
print ('Représentation graphique de la loi binomial')
def binomial(N,n,p):
 s=(n+1)*[0];e=(n+1)*[0]
 for i in range(N):
     c=0
 for j in range(n):
     x=random()
 if x<p:
     c=c+1
     s[c]=s[c]+1
 for i in range(n+1):
     s[i]=s[i]/N
 for i in range(n+1):
     e[i]=binom(n,i)*p**i*(1-p)**(n-i)
     r=range(n+1)
 plt.plot(r,s,'o',color='green')
 plt.plot(r,e,'*',color='red')
 plt.axis([0,n+1,0,1])
 plt.show()
binomial(10,1,2)
```

```python
# simulation de la densité de la loi uniforme
from random import random
print('simulation de la densité de la loi uniforme')
def uniforme(a,b):
    return(a+(b-a)*random())
uniforme (-1,2)
```

```python
#représentation graphique de la loi uniforme
import numpy as np
import matplotlib.pyplot as plt
print ("représentation graphique de la densité de la loi uniforme")
def densuni(x):
#densité pour a=-1 et b=2
    if (-1<=x and x<=2):
        return(1/5)
    else:
        return(0)
abs=np.linspace(-5,6,200)
y=[densuni(x) for x in abs] #à cause des inégalités dans densuni on construit les ordonnées à la main
plt.grid()
plt.plot(abs,y)
plt.show()
```

```python
#fonction de répartition pour a=-1 et b=2
def repartuni(x):
    if x<-1:
        return(0)
    elif (-1<=x and x<=2):
        return((x+1)/1)
    else:
        return(1)

abs=np.linspace(-5,5,200)
y=[repartuni(x) for x in abs]
plt.grid()
plt.plot(abs,y)
plt.show()
repartuni(-0.2)
#−4 −2 0 2 4
#−0.2
#0.0
#0.2
#0.4
#0.
```

```python
#Loi exponentielle
import numpy as np 
import matplotlib.pyplot as plt  
in_array = [1, 1.2, 1.4, 1.6, 1.8, 2] 
out_array = np.exp(in_array)  
y = [1, 1.2, 1.4, 1.6, 1.8, 2] 
plt.plot(in_array, y, color = 'blue', marker = "*", label ="loi empirique")   
# red for numpy.exp() 
plt.plot(out_array, y, color = 'red', marker = "o", label ="fonction de densité") 
plt.title("loi exponentielle") 
plt.xlabel("X", color = "green") 
plt.ylabel("Y", color = "green") 
plt.legend()
plt.show()   
```
